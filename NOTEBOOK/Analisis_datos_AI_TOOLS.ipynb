# **`Top 5 IA más usadas por mes`**

```python
top5 = df.sort_values(by='Monthly_Traffic_Est', ascending=False).head(5)

top5[['Tool_Name','Monthly_Traffic_Est']]

plt.figure(figsize=(8,5))

sns.barplot(
    x='Monthly_Traffic_Est',
    y='Tool_Name',
    data=top5,
    palette='viridis'
)

plt.title("Top 5 herramientas de IA más usadas (tráfico mensual)", fontsize=14)
plt.xlabel("Tráfico mensual estimado")
plt.ylabel("Herramienta de IA")

plt.show()
```

El análisis del tráfico mensual muestra que ChatGPT es la herramienta de inteligencia artificial más utilizada, con aproximadamente 1.8 mil millones de visitas mensuales, superando ampliamente a las demás plataformas. Le siguen Canva AI y Gemini con alrededor de 500 millones de visitas, lo que indica que estas herramientas también tienen una alta adopción por parte de los usuarios. En menor medida aparecen Character.AI y Google Workspace AI, que igualmente mantienen un uso significativo dentro del ecosistema de herramientas de inteligencia artificial.

`**IA más buenas para programación**`

```python
df[df['Primary_Category'].str.contains("Coding", case=False)].sort_values(by='User_Rating', ascending=False)

plt.figure(figsize=(8,5))

sns.barplot(
    x='User_Rating',
    y='Tool_Name',
    data=coding_ai,
    palette='mako'
)

plt.title("Herramientas de IA para programación mejor calificadas")
plt.xlabel("Calificación de usuarios")
plt.ylabel("Herramienta de IA")

plt.show()
```

El análisis de herramientas de inteligencia artificial enfocadas en programación muestra que Cursor y GitHub Copilot se encuentran entre las mejor calificadas por los usuarios, con puntuaciones cercanas a 4.7 y 4.6 respectivamente. También destacan herramientas como Terra Core y Codeium, que mantienen buenas valoraciones dentro de la categoría de asistentes de programación. Esto indica que las herramientas de IA para desarrollo de software están bien valoradas y tienen una adopción importante entre los programadores.

`**IA gratuitas**`

```python
gratis = df[df['Pricing_Model'].str.contains("Free|Freemium", case=False)]

gratis.sort_values(by='User_Rating', ascending=False).head(10)[['Tool_Name','Pricing_Model','User_Rating']]
```

Los resultados muestran que muchas de las IA mejor valoradas por los usuarios son gratuitas o freemium, lo que facilita su adopción masiva y explica por qué tienen una gran popularidad entre estudiantes, desarrolladores y profesionales.

**`IA de paga`**

```python
df[df['Pricing_Model'] == "Paid"][['Tool_Name','Pricing_Model','Starting_Price_USD']]
```

El análisis muestra que muchas herramientas de IA utilizan un modelo de negocio basado en suscripción, donde el precio depende del nivel de funcionalidades, uso profesional o capacidades avanzadas que ofrece la plataforma. Las herramientas más populares suelen mantener precios accesibles para atraer usuarios, mientras que las soluciones más especializadas tienen costos más elevados debido a sus capacidades técnicas o empresariales.

**`Cuál es el precio promedio de las IA de pago?`**

```python
df['Starting_Price_USD'].mean()
```

El precio promedio de las herramientas de inteligencia artificial de pago es de aproximadamente $61.62 USD mensuales. Esto significa que, en general, las plataformas de IA con planes pagos tienen un costo cercano a ese valor, aunque algunas pueden ser más económicas y otras más costosas dependiendo de sus funciones.

`**¿Cuántas IA son gratuitas vs pagas?**`

Esto indica que aunque existen opciones gratuitas o con acceso limitado, la mayoría de las plataformas de IA dependen de modelos de monetización basados en pago o suscripción.
